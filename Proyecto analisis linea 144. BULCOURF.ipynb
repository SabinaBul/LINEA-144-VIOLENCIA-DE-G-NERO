# Definición del problema y modelo

A partir de los hallazgos iniciales, se planea generar un agrupamiento de los distintos conjuntos afectados por la violencia de género, clasificándose según sus características.
Para resolver este problema, se utilizará el tipo de aprendizaje no supervisado y se aplicará el modelo de clusterización. Este modelo es el más apropiado ya que permite un análisis de los datos organizados en clústeres que facilitarán encontrar las similitudes en los conjuntos y revelará información que nos permitirá definir de forma más eficiente los grupos y los aspectos de su afectación por la problemática.


# Codificación de variables

```python
#Pasamos columnas de valores tipo object a category y después esos mismos a numéricos

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype('category')
    df[col] = df[col].cat.codes

#Le pedimos que nos muestre los números asignados a cada valor
print(df)
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
constant_filter = VarianceThreshold(threshold=0)
```

```python
# prompt: eliminar variables de varianza nula

# Aplicar el filtro a las columnas numéricas
df_numeric = df.select_dtypes(include=['number'])
constant_filter.fit(df_numeric)

# Obtener las columnas que NO tienen varianza cero
constant_columns = [col for col in df_numeric.columns if constant_filter.variances_[df_numeric.columns.get_loc(col)] > 0]

# Crear un nuevo DataFrame solo con las columnas con varianza
df_filtered_variance = df[constant_columns].copy()

print("\nColumnas con varianza nula eliminadas:")
print([col for col in df_numeric.columns if col not in constant_columns])
print("\nDataFrame después de eliminar columnas con varianza nula:")
print(df_filtered_variance.head())
print("\nInformación del DataFrame después de eliminar columnas con varianza nula:")
df_filtered_variance.info()
```

# Implementación y entrenamiento del modelo

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
data_scaled= df_filtered_variance.copy()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(data_scaled)
data_scaled= pd.DataFrame(scaler.transform(data_scaled), columns= data_scaled.columns)
```

```python
data_scaled.head()
```

```python
scaledPCA = pca.fit_transform(data_scaled)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
dendogramPlot = dendrogram(linkage(scaledPCA, method = "ward"))
plt.title("Dendograma usando ward linkage")
plt.xlabel("Cluster")
plt.ylabel("Distancia Euclidiana")
plt.show()
```

```python
pca = PCA(n_components=2)
scaledPCA = pca.fit_transform(data_scaled)
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans

Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(scaledPCA)
Elbow_M.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Calcular el score de silueta para diferentes números de clusters
silhouette_scores = []
# Rango de k a probar (de 2 a 9, ya que el codo sugiere 3)
# Evitamos k=1, ya que el score de silueta requiere al menos 2 clusters
range_n_clusters = range(2, 10)

for n_clusters in range_n_clusters:
    # Inicializar KMeans con n_clusters
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10) # n_init para evitar warnings
    cluster_labels = kmeans.fit_predict(scaledPCA)

    # Calcular el score de silueta
    silhouette_avg = silhouette_score(scaledPCA, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"Para n_clusters = {n_clusters}, el score de silueta es: {silhouette_avg}")

# Graficar el score de silueta para cada k
plt.figure(figsize=(8, 6))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title('Método del Score de Silueta para Determinar el Número Óptimo de Clusters')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Score de Silueta Promedio')
plt.xticks(range_n_clusters)
plt.grid(True)
plt.show()
```

**Ambos métodos coinciden en que el número óptimo de clústeres es 3. Por lo que utilizaremos ese número.**

Luego de correr todo de nuevo, el número de silueta a veces se modifica a 2, pero es inestable. Decidimos tomar la decisión de quedarnos con el número 3. Además, porque ya se probó el modelo con 2 clusteres y se vuelve muy complejo el análisis porque los grupos no quedan bien definidos.

# Rendimiento del modelo

```python
# prompt: graficar los tres clusters

import matplotlib.pyplot as plt
# Ajustar K-Means con 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(scaledPCA)

# Agregar las etiquetas de cluster al DataFrame original o al DataFrame PCA escalado
data_scaled['Cluster'] = clusters

# Visualizar los clusters usando los primeros 2 componentes principales
plt.figure(figsize=(10, 8))
sns.scatterplot(x=scaledPCA[:, 0], y=scaledPCA[:, 1], hue=clusters, palette='viridis', s=50, alpha=0.6)
plt.title('Visualización de Clusters (PCA de 2 Componentes)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# Opcional: Visualizar los centros de los clusters
centers = kmeans.cluster_centers_
plt.figure(figsize=(10, 8))
sns.scatterplot(x=scaledPCA[:, 0], y=scaledPCA[:, 1], hue=clusters, palette='viridis', s=50, alpha=0.6)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.8, marker='X', label='Centros de Cluster')
plt.title('Visualización de Clusters y Centros (PCA de 2 Componentes)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

Según estos gráficos, hay tres grupos bien diferenciados.

# Conclusiones



**El clúster 0:** representa al grupo con mayor **diversidad etaria.** Nuclea (además de a muchas mujeres) un gran grupo de casos de identidades de la **diversidad sexual,** y varones. Este cluster representa los casos de violencia que se dan fuera del ámbito de la pareja, y en el que predominan los tipos de violencia psicológica, simbólica y sexual. El clúster 0 tiene un perfil diverso y **extra-matrimonial.**

**El clúster 1:**  este clúster representa mayoritariamente a adultos y adultos mayores, y al haber una fuerte presencia de mujeres (aunque haya una presencia menor de varones y personas no binarias) puede inducirse que este clúster nuclea a **las mujeres adultas y mayores.** Al haber una mayor presencia de mujeres, aumenta la presencia del tipo de violencia doméstica. Los tres tipos de violencia que afectan mayoritariamente a este clúster son: **psicológica, doméstica y simbólica.** Este conjunto representa las situaciones de violencia en y/o post  relaciones afectivas **íntimas.**  Es un clúster de pérfil **íntimo, feminizado, y con tendencia a adultas y mayores**.



**El clúster 2:** en el cluster 2 hay una predominancia casi total de **mujeres**, con el resto de los géneros en un número muy bajo. Representa un rango etario un poco menor, podría decirse que el clúster nuclea a las **mujeres jóvenes y adultas** predominantemente. Muestra a este grupo de mujeres jóvenes más afectado por un patrón parecido al clúster 1, pero en mayor escala, en el que los vínculos con el agresor son de **ex-pareja o pareja.**  Este grupo se ve más afectado por la violencia psicológica, doméstica y física. Por lo que puede suponerse que es un grupo más expuesto a situaciones dentro del **hogar, y en la pareja**, que implican extorsión y hasta **agresión física.**  Este clúster tiene un perfil: **joven, feminizado,** y de tipo de agresión doméstica-física.

En todos los clústeres hay un grupo de género que predomina, y son **las mujeres**. En todos los clústeres el 50% tiene **34 años, por lo que podemos concluir que es la edad más afectada**, según estos casos tomados, por la problemática. Por lo que será necesario implementar propuestas pensadas para **mujeres jóvenes-adultas.**
La violencia ejercida por **parejas y ex-parejas** domina ampliamente los casos en 2 de los 3 clústeres.
En los tres clústers el tipo de violencia que más afecta a los grupos es el de violencia **psicológica.** Por lo que puede decirse que este tipo de violencia afecta a todos los **géneros transversalmente** por igual.
